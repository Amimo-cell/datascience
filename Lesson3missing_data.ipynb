```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import io
from google.colab import files
```

```python
# allow us to import data
data= files.upload()
```

```python
#converting the file into a dataframe
sample_submission = pd.read_csv(io.StringIO(data['Diabetes Missing Data.csv'].decode('utf-8')))
```

```python
#view the dataframe
sample_submission.head()
```

```python
# Generating data info
sample_submission.info()
```

```python
# Getting the missing data
sample_submission.isnull().sum()
```

```python
#drop missing values in Glucose
sample_submission.dropna(subset=["Glucose"],axis=0, inplace=True)
```

```python
sample_submission.isnull().sum()
```

```python
# Replace Null values in Diastolic_BP with zero
sample_submission["Diastolic_BP"]=sample_submission["Diastolic_BP"].fillna(0)
sample_submission.isnull().sum()
```

```python
# calculate the mean for skin_fold values
mean=sample_submission['Skin_Fold'].mean()
```

```python
# Replace Null values in Skin_Fold with average value
sample_submission["Skin_Fold"].replace(np.NAN,mean)
sample_submission.isnull().sum()
```

```python
#replace null values in Skin_Fold with average value
sample_submission['Skin_Fold'].fillna(sample_submission['Skin_Fold'].mean(), inplace=True)
sample_submission.isnull().sum()
```
