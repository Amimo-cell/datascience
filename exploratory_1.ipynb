```python
import pandas as pd
import io
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
data = files.upload()
```

```python
#converting the file into a dataframe
data= pd.read_csv(io.StringIO(data['car_price_dataset (1).csv'].decode('utf-8')))
#view the dataframe
data.head()
```

```python
#Group by Brand, Model, Price
data_test=data[['Brand','Model', 'Price']]
data_grp= data_test.groupby(['Brand', 'Model'], as_index=False).mean()
data_grp
```

```python
#Pivot Table
data_pivot=data_grp.pivot(index='Brand', columns='Model')
data_pivot
```

```python
data_test=data[['Brand','Model' 'Price']]
data_grp= data_test.groupby(['Brand'], as_index=False).mean()
data_grp
```

```python
#Group By Brand, Fuel_Type and Price
data_test=data[['Brand','Fuel_Type', 'Price']]
data_grp= data_test.groupby(['Brand', 'Fuel_Type'], as_index=False).mean()
data_grp
```

```python
#pivot table
data_pivot=data_grp.pivot(index='Brand', columns='Fuel_Type')
data_pivot
```

```python
#Plot target variable over multiple variables
plt.pcolor(data_pivot, cmap='RdBu')
plt.colorbar()
plt.xlabel('Fuel_Type')
plt.ylabel('Brand')
plt.show()
```
