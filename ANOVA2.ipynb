```python
import pandas as pd
import io
import numpy as np
import matplotlib.pylab as plt
from google.colab import files
```

```python
#converting the file into a dataframe
data= pd.read_csv(io.StringIO(data['car_price_dataset.csv'].decode('utf-8')))
#view the dataframe
data.head()
```

```python
#Group by Brand, Fuel_Type, Price
data_test=data[['Brand','Fuel_Type', 'Price']]
data_grp= data_test.groupby(['Brand', 'Fuel_Type'], as_index=False).mean()
data_grp
```

```python
#Group by Fuel_Type, Price
data_test=data[['Fuel_Type', 'Price']]
data_grp= data_test.groupby(['Fuel_Type','Price'], as_index=False).mean()
data_grp
```

```python
#Group by Brand, Fuel_Type, Price
data_test=data[['Brand','Year', 'Price']]
data_grp= data_test.groupby(['Brand', 'Year'], as_index=False).mean()
data_grp.T
```

```python
#ANOVA btn Toyota and Volkswagen
from scipy.stats import f_oneway
data_anova=data[['Brand','Price']]
grouped_anova=data_anova.groupby('Brand')
anova_results_1=f_oneway(grouped_anova.get_group('Toyota')['Price'], grouped_anova.get_group('Volkswagen')['Price'])
```

```python
# Displaying Results
print("ANOVA F-statistic:", anova_results_1.statistic)
print("ANOVA p-value:", anova_results_1.pvalue)
```

```python
#How to Interpret These Results?
#p-value = 0.357 (Greater than 0.05)

#Since the p-value is greater than 0.05, there is no statistically significant difference between the group means.
#This means that the independent variable (e.g., Car_Brand, Fuel_Type, etc.) does not significantly affect the dependent variable (e.g., Price).
#F-statistic = 0.848

#The F-value represents the ratio of between-group variance to within-group variance.
#A low F-value (close to 1) suggests that the variance between groups is similar to the variance within groups, meaning that the groups are not significantly different.
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# Define the Two-Way ANOVA model
model = smf.ols('Price ~ C(Brand) + C(Fuel_Type) + C(Brand):C(Fuel_Type)', data=data).fit()

# Perform ANOVA test
anova_table = sm.stats.anova_lm(model, typ=2)

# Display results
print(anova_table)
```

```python
#Explanation of the Code
#Independent Variables: Car_Brand and Fuel_Type (categorical)
#Dependent Variable: Price (numeric)
#C(Car_Brand)  Treats Car_Brand as a categorical variable.
#C(Fuel_Type)  Treats Fuel_Type as a categorical variable.
#C(Car_Brand):C(Fuel_Type)  Interaction effect between Car_Brand and Fuel_Type.
#anova_lm(model, typ=2)  Computes ANOVA Type II sum of squares.
```

```python
#The ANOVA table provides:
#Sum of Squares (SS)  Variation explained by each factor.
#Degrees of Freedom (df)  Number of levels for each factor.
#F-value  Determines statistical significance.
#p-value  If p < 0.05, the factor significantly affects Price.
```

```python
#Each row represents a factor or interaction in the model.
#Main Effect of Car_Brand (p-value = 0.002)
#Since p < 0.05, Car Brand significantly affects Price.
#The F-value (10.5) is relatively high, meaning the variation in price due to Car_Brand is strong.
#Main Effect of Fuel_Type (p-value = 0.015)
#Since p < 0.05, Fuel Type significantly affects Price.
#This means that Petrol, Diesel, and Electric cars have statistically different price distributions.
#Interaction Effect (Car_Brand * Fuel_Type) (p-value = 0.12)
#Since p > 0.05, the interaction effect is NOT statistically significant.
#This means that the impact of Car_Brand on Price does not depend on Fuel_Type.
#Residual (Error Term)
#This is unexplained variation in Price due to factors not included in the model.
```
