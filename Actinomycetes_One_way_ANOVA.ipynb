```python
import pandas as pd
import io
import numpy as np
import matplotlib.pylab as plt
from scipy.stats import f_oneway
from google.colab import files
```

```python
data = files.upload()
```

```python
#converting the file into a dataframe
data= pd.read_csv(io.StringIO(data['Actinomycetes.final (1).csv'].decode('utf-8')))
#view the dataframe
data.head()
```

```python
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Clean column names
data.columns = data.columns.str.strip()

# Reshape data to long format
data_melted = data.melt(id_vars=["Antibiotics"], var_name="Isolate", value_name="Zone_Diameter")

data.head()
```

```python
#Null hypotheses (H₀):There a no significant difference in the inhibition zone among different antibiotics
#Alternative hypotheses (H₁):There is a statistically significant difference in the inhibition zone among the antibiotics.
```

```python
# Perform one-way ANOVA for antibiotics
model = ols('Zone_Diameter ~ C(Antibiotics)', data=data_melted).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
# Conclusion: Reject the null hypothesis
#The antibiotics have a significant effect on the inhibition zone
```

```python
#Tukey's Post-Hoc Analysis
tukey = pairwise_tukeyhsd(endog=data_melted['Zone_Diameter'], groups=data_melted['Antibiotics'], alpha=0.05)
print(tukey)
```
