```python
data['Temperature (°C)'] = data['Temperature (°C)'].astype(int)
```

```python
data['Ice Cream Sales (units)']= data['Ice Cream Sales (units)'].astype(int)
```

```python
data.columns = data.columns.str.strip()
```

```python
data.rename(columns={'Temperature (°C)': 'Temperature'}, inplace=True)
data.rename(columns={'Ice Cream Sales (units)': 'Ice_cream_Sales'}, inplace=True)
```

```python
SCALE = StandardScaler()
SCALE.fit(data[['Temperature', 'Ice_cream_Sales']])
scaled_data = SCALE.transform(data[['Temperature', 'Ice_cream_Sales']])
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression # Assuming you want to use Linear Regression

# Define the pipeline steps
steps = [('scaler', StandardScaler()), ('regressor', LinearRegression())]

# Create the pipeline
Pipe = Pipeline(steps)

# Now you can use Pipe.fit and Pipe.predict
Pipe.fit(data[['Temperature']], data['Ice_cream_Sales']) # Assuming 'Ice_cream_Sales' is your target variable
Yhat = Pipe.predict(data[['Temperature']]) # Predicting on the same data for demonstration
```
