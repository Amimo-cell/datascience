```python
# converting the data into a data frame
from io import StringIO
import urllib.request

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

# URL to a CSV file
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data/"
# Replace with your actual URL

# Fetch the content from the URL
with urllib.request.urlopen(url) as response:
    data = response.read().decode('utf-8')

# Convert the content into a DataFrame
df = pd.read_csv(StringIO(data) ,names=headers, header=None)

# Display the DataFrame
print(df.T)
```

```python
# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head()
```

```python
for column in df.columns.values.tolist():
    print(column)
    print (df[column].value_counts())
    print("")
```

```python
# Calculate percentage of missing values for each column
missing_percentage = df.isnull().sum() * 100 / len(df)

print(missing_percentage)
```

```python
#Calculate the mean value for the "normalized-losses" column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
#Replace "NaN" with mean value in "normalized-losses" column
avg_norm_loss = df["normalized-losses"].astype("int64").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
#Replace "NaN" with mean value in "normalized-losses" column
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
df["normalized-losses"].isnull().sum()
```

```python
#Calculate the mean value for the "bore" column
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
#Replace "NaN" with the mean value in the "bore" column
df["bore"].replace(np.nan, avg_bore, inplace=True)
df["bore"].isnull().sum()
```

```python
#Calculate the mean value for the "stroke" column
avg_stroke=df['stroke'].astype('float').mean(axis=0)
print("Average of stroke:", avg_stroke)
```

```python
#Replace "NaN" with the mean value in the "stroke" column
df["stroke"].replace(np.nan, avg_stroke, inplace=True)
df["stroke"].isnull().sum()
```

```python
#values are present in a particular column
df['num-of-doors'].value_counts()
```

```python
#values are present in a particular column
df['num-of-doors'].value_counts().idxmax()
```

```python
#replace the missing 'num-of-doors' values by the most frequent
df["num-of-doors"].replace(np.nan, "4", inplace=True)
df["num-of-doors"].isnull().sum()
```

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
df["price"].isnull().sum()
```
