```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
import seaborn as sns
from google.colab import files
```

```python
data=files.upload()
```

```python
#converting the file into a dataframe
data= pd.read_csv(io.StringIO(data['car_price_dataset (1).csv'].decode('utf-8')))
#view the dataframe
data.head()
```

```python
data.head().T
```

```python
# Calculating the pearson coefficient
import scipy.stats as stats
pearson_coef, p_value=stats.pearsonr(data['Engine_Size'], data['Price'])
print(pearson_coef)
print(p_value)
```

```python
#interpreting our results
#The value of 0.3574 indicates a weak to moderate positive linear relationship between the two variables
#The p-value is extremely small (much less than 0.05), which indicates that the correlation is statistically significant.
```

```python
#P. value of < 0.001= strong certainty
#p.Value of > 0.1 =No certainty in the results
#Pearson_coef ( close to +1= Large positive relationship)
#Pearson_coef ( close to -1= Large negative relationship, 0 =No relationship)
```

```python
# Select specific columns
selected_columns = ['Engine_Size', 'Price',]  # Modify this list to include only the desired columns
df_selected = data[selected_columns]
```

```python
# Compute correlation matrix for selected columns
corr_matrix = df_selected.corr()
```

```python
# Create heatmap
plt.figure(figsize=(5, 4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap (Selected Columns)")
plt.show()



```
