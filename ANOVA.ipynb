```python
import pandas as pd
import io
import numpy as np
import matplotlib.pylab as plt
from google.colab import files
```

```python
data = files.upload()
```

```python
#converting the file into a dataframe
data= pd.read_csv(io.StringIO(data['car_price_dataset.csv'].decode('utf-8')))
#view the dataframe
data.head()
```

```python
#Group by Brand, Model, Price
data_test=data[['Brand','Model', 'Price']]
data_grp= data_test.groupby(['Brand', 'Model'], as_index=False).mean()
data_grp
```

```python
#ANOVA btn Audi and BMW
from scipy.stats import f_oneway
data_anova=data[['Brand','Price']]
grouped_anova=data_anova.groupby('Brand')
anova_results_1=f_oneway(grouped_anova.get_group('Audi')['Price'], grouped_anova.get_group('BMW')['Price'])

```

```python
# Displaying Results
print("ANOVA F-statistic:", anova_results_1.statistic)
print("ANOVA p-value:", anova_results_1.pvalue)
```

```python
#Since your p-value (0.103) > 0.05, this means:
#The F-value (2.655) is not high enough to indicate a statistically significant difference.
# The observed variance between Audi and BMW could be due to random chance rather than a true price difference.
```

```python
#The results show minimal statistical significant differences.
#p-value	Interpretation
#p  0.05	Statistically significant difference
#p > 0.05	Not statistically significant
#p  0.10	Weak evidence for a difference (not conclusive)
#The F-statistic is calculated as:
#= Variancebetweengroups/Variancewithingroups
#Higher F-value  Greater difference between groups.
#Lower F-value  Groups are more similar.

```
